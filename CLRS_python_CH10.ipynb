class TStack(list):

class Queue(list):
    def __init__(self,n,head):
        self.length = n
        self[:] = [float('inf') for _ in range(n)]
        self.head = self.tail=head-1
    
    def enqueue(self,x):
        if self.head == self.tail+1:
            raise Exception("stack overflow!")
        if self.tail!=self.length-1:
            self.tail += 1
        else:
            self.tail = 0
        self[self.tail-1]=x
    
    def dequeue(self):
        if self.head==self.tail:
            raise Exception("stack underflow!")
        if self.head==self.length-1:
            self.head = 0
            return self[-1]
        else:
            self.head += 1
            return self[self.head-1]
